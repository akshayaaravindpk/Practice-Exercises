```python
df.drop(['Id','Location'],axis=1,inplace=True)
```

```python
df['Target']=df['Target'].map({'Positive':1,'Negative':-1,'Neutral':0})
```

```python
df
```

```python
tweets=df.Text
tweets
```

```python
#remove special characters
tweets=tweets.str.replace('[^a-zA-Z0-9]+'," ")
tweets
```

```python
#stemming
import nltk
from nltk.stem import SnowballStemmer
from nltk import TweetTokenizer
stemmer=SnowballStemmer('english')
tk=TweetTokenizer()
```

```python
tweets=tweets.apply(lambda x:[stemmer.stem(i.lower()) for i in tk.tokenize(x)]).apply(lambda y:" ".join(y))
tweets
```

```python
#removing stopwords
from nltk.corpus import stopwords
nltk.download('stopwords')
sw=stopwords.words('english')
print(sw)
```

```python
tweets=tweets.apply(lambda x:[i for i in tk.tokenize(x) if i not in sw]).apply(lambda y:" ".join(y))
tweets
```

```python
#vectorization
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
train_vec=vectorizer.fit_transform(tweets)
print(train_vec)
```

```python
print(train_vec.shape)
```

```python
X=train_vec
y=df['Target'].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
sv=SVC()
mn=MultinomialNB()
rf=RandomForestClassifier()
ab=AdaBoostClassifier()
lst=[sv,mn,rf,ab]
```

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay
for i in lst:
  print(i)
  i.fit(X_train,y_train)
  y_pred=i.predict(X_test)
  s="facebook is a hub of fake information"
  y_new=i.predict(vectorizer.transform([s]))
  print("******************************-PREDICTION FOR NEW STRING-******************************")
  if y_new==1:
    print("***************************************-POSITIVE-**************************************")
  elif y_new==0:
    print("***************************************-NEUTRAL-***************************************")
  else:
    print("***************************************-NEGATIVE-**************************************")
  print(classification_report(y_test,y_pred))
  print("_______________________________________________________________________________________")
  print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```
