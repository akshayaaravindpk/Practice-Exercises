```python
flat_data=[]
target=[]
Categories=['Cat','Dog']
datadir='/content/drive/MyDrive/Data'
```

```python
import numpy as np
for i in Categories:                            #Cat or dog
  print('loading...Categories',i)
  path=os.path.join(datadir,i)                  #/content/drive/MyDrive/Data/Cat
  for img in os.listdir(path):
    img_array=imread(os.path.join(path,img))    #/content/drive/MyDrive/Data/Cat/cat.1 - Copy.jpg
    img_resized=resize(img_array,(150,150,3))
    flat_data.append(img_resized.flatten())
    target.append(Categories.index(i))          #0,1
  print("loaded Category:",i,"Successfull")
```

```python
img_array
```

```python
df=pd.DataFrame(flat_data)
df
```

```python
img_resized
```

```python
df['target']=target
df
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=77)
```

```python
from sklearn.svm import SVC
sv=SVC()
sv.fit(x_train,y_train)
y_pred=sv.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```
